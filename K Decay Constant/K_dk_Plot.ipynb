```python
"""
Upper Bound Data
"""

c_hb_ar = df['Carter_hb_ar'].values
min_hb_ar = df['Min_hb_ar'].values
fra_diff_hb_ar = df['fra_diff_hb_ar'].values
diff_hb_ar = df['diff_hb_ar'].values

c_hb_kar = df['carter_hb_kar'].values
min_hb_kar = df['min_hb_kar'].values
fra_diff_hb_kar = df['fra_diff_hb_kar'].values
diff_hb_kar = df['diff_hb_kar'].values
```

```python
sns.set_style('ticks')
font = {"family": "Arial", 
       "weight": "normal", 
       "size": 16}

plt.rc("font", **font)

linestyle_tuple = [
     ('loosely dotted',        (0, (1, 10))),
     ('dotted',                (0, (1, 1))),
     ('densely dotted',        (0, (1, 1))),

     ('loosely dashed',        (0, (5, 10))),
     ('dashed',                (0, (5, 5))),
     ('densely dashed',        (0, (5, 1))),

     ('loosely dashdotted',    (0, (3, 10, 1, 10))),
     ('dashdotted',            (0, (3, 5, 1, 5))),
     ('densely dashdotted',    (0, (3, 1, 1, 1))),

     ('dashdotdotted',         (0, (3, 5, 1, 5, 1, 5))),
     ('loosely dashdotdotted', (0, (3, 10, 1, 10, 1, 10))),
     ('densely dashdotdotted', (0, (3, 1, 1, 1, 1, 1)))]
```

```python
fig= plt.figure(figsize = (6,6))
plt.plot(c_hb_ar, diff_hb_ar, 
        label = 'Ar-Ar EC/$\\beta$= 212', 
        lw = 3, 
        color = 'k')
plt.plot(c_hb_ar, diff_hb_kar, 
        label = 'K-Ar EC/$\\beta$= 212', 
        lw = 3, 
        color = 'k', 
        ls = '--')
plt.plot(c_hb_ar, diff_lb_ar, 
        label = 'Ar-Ar EC/$\\beta$ = 150', 
        lw = 3, 
        ls = '--', 
        color = 'grey')
plt.plot(c_hb_ar, diff_lb_kar, 
        label = 'K-Ar EC/$\\beta$ = 150', 
        lw = 3, 
        ls = 'dashdot', 
        color = 'grey')
plt.xlim(0, 5000)
plt.xlabel('Age (Ma)')
plt.ylabel('$\Delta$Age (Ma)')
plt.legend(fontsize= 12)
```

```python
fig, ax = plt.subplots(1, 2, figsize = (10, 5))

ax[0].plot(c_hb_ar, diff_lb_ar, 
        lw = 3, label = 'Ar-Ar',
        ls = '-', 
        color = 'grey')
ax[0].plot(c_hb_ar, diff_hb_ar, 
        label = 'Ar-Ar', 
        lw = 3, 
        color = 'k')
ax[1].plot(c_hb_ar, diff_lb_kar, 
        label = 'EC/$\\beta^{+}_{ground}$= 150', 
        lw = 3, 
        color = 'grey')

ax[1].plot(c_hb_ar, diff_hb_kar, 
        label = 'EC/$\\beta^{+}_{ground}$= 212', 
        lw = 3, 
        color = 'k', 
        ls = '-')





ax[0].set_ylabel('$\Delta$Age (Ma)')
ax[1].set_xlabel('Age (Ma)')
ax[0].set_xlim(0, 5000)
ax[1].set_xlim(0, 5000)
ax[1].set_ylim(-70, 0)
plt.legend(fontsize = 14)


ax3 =plt.axes([0.21,0.21,0.18,0.25])

ax3.plot(c_hb_ar, fra_diff_hb_ar*100, 
        color = 'k', 
        ls = '-')
ax3.plot(c_hb_ar, fra_diff_lb_ar*100, 
        color = 'grey', 
        ls = '-')
ax3.tick_params(labelsize = 9)
ax3.locator_params(axis='x', nbins=6)
ax3.locator_params(axis='y', nbins=6)

ax3.set_xlim(0, 5000)
ax3.set_ylabel('Fractional Difference (%)', 
              fontsize= 10)
ax3.set_xlabel('Age (Ma)', 
              fontsize= 10)


ax4=plt.axes([0.63,0.21,0.18,0.25])

ax4.plot(c_hb_ar, fra_diff_hb_kar*100, 
        color = 'k', 
        ls = '-')
ax4.plot(c_hb_ar, fra_diff_lb_kar*100, 
        color = 'grey', 
        ls = '-')
ax4.tick_params(labelsize = 9)
ax4.locator_params(axis='x', nbins=6)
ax4.locator_params(axis='y', nbins=6)

ax4.set_xlim(0, 5000)
ax4.set_ylabel('Fractional Difference (%)', 
              fontsize= 10)

ax4.set_xlabel('Age (Ma)', 
              fontsize= 10)

plt.savefig('Delta_age_Figure.pdf')
```
