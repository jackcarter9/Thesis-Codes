```python
# DATA 
import pandas as pd
cwd = os.chdir('/Users/jackcarter/Missoula/Missoula Data/Missoula Data Final /Moses Coulee')
dfa1 = pd.read_csv("a_b1_data.csv")
dfa2 = pd.read_csv("b_b1_data_y.csv")
dfa3 = pd.read_csv("e_b1_data_y.csv")
dfa4 = pd.read_csv("f_b1_data_y.csv")
age_mc1 = dfa1['Age'].values
age_mc2 = dfa2['Age'].values
age_mc3 = dfa3['Age'].values
age_mc4 = dfa4['Age'].values
```

```python
from scipy.stats import gaussian_kde
def kde_plot(x, grid, bandwidth = 20):
    y = gaussian_kde(x, bw_method = bandwidth / x.std(ddof =1))
    return y.evaluate(grid)


grid = np.linspace(0, 4600, 1000)

age_ef1_kde = kde_plot(age_ef1, grid, bandwidth = 10)
age_ef2_kde = kde_plot(age_ef2, grid, bandwidth = 10)
age_ef3_kde = kde_plot(age_ef3, grid, bandwidth = 10)

age_mc1_kde = kde_plot(age_mc1, grid, bandwidth = 10)
age_mc2_kde = kde_plot(age_mc2, grid, bandwidth = 10)
age_mc3_kde = kde_plot(age_mc3, grid, bandwidth = 10)
age_mc4_kde = kde_plot(age_mc4, grid, bandwidth = 10)
```

```python
font = {"family": "Arial", 
       "weight": "normal", 
       "size": 12}

plt.rc("font", **font)
sns.set_style('ticks')

fig, ax = plt.subplots(1, 1, figsize= (8, 2))
"""
Plot
"""

"""
Ephrata Fan
"""
ax.fill_between(grid, age_ef1_kde/age_ef1_kde.sum(), 
                  label = '18-IAF-10S-EPH', 
                  color = 'steelblue',
               alpha = 0.6)
ax.fill_between(grid, age_ef2_kde/age_ef2_kde.sum(), 
                  label = '18-IAF-09S-EPH', 
                  color = 'seagreen',
               alpha = 0.6)
ax.fill_between(grid, age_ef3_kde/age_ef3_kde.sum(), 
                  label = '18-IAF-10S-EPH (Silt)', 
                  color = 'orange',
               alpha = 0.6)



ax.plot(grid, age_ef1_kde/age_ef1_kde.sum(), 
          lw = 0.5,
               alpha = 0.6)
ax.plot(grid, age_ef2_kde/age_ef2_kde.sum(), 
          lw = 0.5,
               alpha = 0.6)
ax.plot(grid, age_ef3_kde/age_ef3_kde.sum(), 
          lw = 0.5,
               alpha = 0.6)

"""
Moses Coulee
"""
ax.fill_between(grid, age_mc1_kde/age_mc1_kde.sum(), 
                  label = '18-IAF-05-MOS', 
                  color = 'r', 
                  alpha = 0.6)
ax.fill_between(grid, age_mc2_kde/age_mc2_kde.sum(), 
                  label = '18-IAF-06-MOS', 
                  color = 'b', 
                  alpha = 0.6)
ax.fill_between(grid, age_mc3_kde/age_mc3_kde.sum(), 
                  label = '18-IAF-05-MOS (Silt)', 
                  color = 'cyan', 
                  alpha = 0.6)
ax.fill_between(grid, age_mc4_kde/age_mc4_kde.sum(), 
                  label = '18-IAF-06-MOS (Silt)', 
                  color = 'pink', 
                  alpha = 0.6)


ax.plot(grid, age_mc1_kde/age_mc1_kde.sum(), 
          lw = 0.5)
ax.plot(grid, age_mc2_kde/age_mc2_kde.sum(), 
          lw = 0.5)
ax.plot(grid, age_mc3_kde/age_mc3_kde.sum(), 
          lw = 0.5)
ax.plot(grid, age_mc4_kde/age_mc4_kde.sum(), 
          lw = 0.5)

"""
Plot layout
"""
ax.set_xlim(0, 1700)
ax.set_ylim(0, 0.1)

"""
axis labels
"""

ax.set_xlabel('Age (Ma)')
ax.set_ylabel('Normalised probability density')

ax.legend(fontsize = 10)
```
