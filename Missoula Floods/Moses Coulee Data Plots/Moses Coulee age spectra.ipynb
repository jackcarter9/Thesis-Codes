```python
age1 = dfa1['Age'].values
age2 = dfa2['Age'].values
age3 = dfa3['Age'].values
age4 = dfa4['Age'].values

age1_er = dfa1['Age_Er'].values
age2_er = dfa2['Age_Er'].values
age3_er = dfa3['Age_Er'].values
age4_er = dfa4['Age_Er'].values

ar391 = dfa1['Ar39_'].values
ar392 = dfa2['Ar39_'].values
ar393 = dfa3['Ar39_'].values
ar394 = dfa4['Ar39_'].values
```

```python
def cumrel(x): 
    y = np.cumsum(x)
    z = y/y.max()
    return z
```

```python
crel1 = cumrel(ar391)
crel2 = cumrel(ar392)
crel3 = cumrel(ar393)
crel4 = cumrel(ar394)
```

```python
# Plotting parameters Data

age_p1 = np.insert(age1,
                     0, age1[0])
age_p_err1 =  np.insert(age1_er, 
                         0, age1_er[0])
ar39_plot_1 = np.insert(crel1,
                      0, 0)

age_p2 = np.insert(age2,
                     0, age2[0])
age_p_err2 =  np.insert(age2_er, 
                         0, age2_er[0])
ar39_plot_2 = np.insert(crel2,
                      0, 0)


age_p3 = np.insert(age3,
                     0, age3[0])
age_p_err3 =  np.insert(age3_er, 
                         0, age3_er[0])
ar39_plot_3 = np.insert(crel3,
                      0, 0)


age_p4 = np.insert(age4,
                     0, age4[0])
age_p_err4 =  np.insert(age4_er, 
                         0, age4_er[0])
ar39_plot_4 = np.insert(crel4,
                      0, 0)
```
