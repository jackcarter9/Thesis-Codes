```python
from scipy.stats import gaussian_kde
def kde_plot(x, grid, bandwidth = 20):
    y = gaussian_kde(x, bw_method = bandwidth / x.std(ddof =1))
    return y.evaluate(grid)


grid = np.linspace(0, 4600, 1000)
```

```python
sns.set_style('ticks')

font= {"family": "Arial", 
      "weight": "normal",
      "size":15}

plt.rc("font", **font)

fig, ax = plt.subplots(1, 1, figsize= (9, 3))
"""
Plot
"""
#############################################
"""
Data
"""


ax.plot(grid, kde_plot(age1, grid, bandwidth = 5)/kde_plot(age1, grid, bandwidth = 5).sum(), 
                alpha = 1, color = 'k', 
               label= '18-IAF-10S-EPH')


"""
Model
"""


for i in range(12):
    

    ax.fill_between(grid, 
        kde_plot(df_age_comp1[f'age{i+1}'].values, grid, bandwidth = 20)/kde_plot(df_age_comp1[f'age{i+1}'], grid, bandwidth = 20).sum(), 
                  alpha = 0.6, 
                   label = f'Age component {i+1}')




"""
Plot layout
"""
#ax.set_xlim(-100, 2300)

ax.set_ylim(0, 0.8)


"""
axis labels
"""

ax.set_xlabel('Age (Ma)')
ax.set_ylabel('Normalised probability density')

ax.legend(ncol = 4,
             fontsize = 10)




# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.locator_params(axis='x', nbins=10)
plt.tight_layout()
ax.set_xlim(0, 1600)
ax.set_ylim(0, 0.2)
plt.savefig('/Users/jackcarter/Missoula/Missoula outputs/mc_ab1_kde_out.pdf')
```
